```python
hulu_df = pd.read_csv('hulu_titles.csv')
```

```python
hulu = hulu_df.copy()
```

```python
def null_rate_fn(df):
    for i in df.columns:
        null_rate = df[i].isna().sum()/len(df) *100
        if null_rate > 0 :
            print('{} null rate is {}%'.format(i, round(null_rate,2)))
```

```python
null_rate_fn(hulu)
```

```python
hulu.drop(columns = ['director','cast'], axis = 1, inplace = True)
```
